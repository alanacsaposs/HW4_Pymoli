### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file_to_load = 'purchase_data.csv'

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

#Display .head
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Calculate the Total Players, Export in DataFrame
total_players = purchase_data['SN'].value_counts().count()
total_players_df = pd.DataFrame({
     'Total Players':[purchase_data['SN'].value_counts().count()]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate:

#Number of Unique Items
Item_List = len(purchase_data['Item ID'].unique())

#Average Purchase Price
Avg_Purchase_Price = purchase_data["Price"].mean()

#Total Number of Purchases                
Total_Purchases = len(purchase_data['Price'])
#Total Revenue
Total_Revenue = purchase_data["Price"].sum()
```

```python
#Display Summary Data Frame
Purchasing_Analysis = pd.DataFrame({'Number of Unique Items': [Item_List], 
                                    'Average Purchase Price': [Avg_Purchase_Price],
                                    'Total Purchases' : [Total_Purchases],
                                    'Total Revenue' : [Total_Revenue]})
Purchasing_Analysis['Average Purchase Price'] = Purchasing_Analysis['Average Purchase Price'].map('${:,.2f}'.format)
Purchasing_Analysis['Total Revenue'] = Purchasing_Analysis['Total Revenue'].map('${:,.2f}'.format)

Purchasing_Analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#See what the breakdown of age looks like
purchase_data['Age'].min()
```

```python
purchase_data['Age'].max()
```

```python
#Create bins
bins = [0,10,14,19,24,29,34,39,50]
group_names=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
#Add New Column to Purchase Data to GroupBy
purchase_data['Age Groups'] = pd.cut(purchase_data['Age'], bins, labels = group_names)
#New DF made with GroupBy Age Groups
age_df = purchase_data.groupby(['Age Groups'])
age_df.head()
```

```python
#Create DF of Age Counts for Calculations
age_count = age_df['SN'].nunique()
age_count = age_count.to_frame()

#Rename 'SN' Column
age_count = age_count.rename(columns={'SN':'Total Count'})
age_count
```

```python
#Add Percentage Column
age_count['Percentage of Players'] = age_count['Total Count'] / total_players * 100
age_count
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create DF with Purchase Data grouped by Item ID and Item Name
pop_items = purchase_data.groupby(['Item ID','Item Name'])
```

```python
#Calculate Purchase Count
item_purchase_count = pop_items['Purchase ID'].count()
#Calculate Item Price
item_price = pop_items['Price'].mean()
#Calculate Total Purchase Value
item_total_value = pop_items['Price'].sum()

#Create Summary Table
item_sum_table = pd.DataFrame({'Purchase Count': item_purchase_count,
                              'Item Price' : item_price,
                              'Total Purchase Value' : item_total_value,
                             })
#Sort by Purchase Count
item_sum_table.sort_values(by='Purchase Count', ascending=False).head()
```

## Most Profitable Items
